```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
text = "data science machine learning artificial intelligence"
wordcloud = WordCloud(width=800, height=400, background_color='pink', colormap='viridis').generate(text)
output_image_path = "wordcloud.png"
wordcloud.to_file(output_image_path)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("WordCloud of Keywords", fontsize=16)
plt.show()
print(f"WordCloud saved as '{output_image_path}'")
```
