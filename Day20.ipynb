```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
def generate_wordcloud(text, output_filename):
    """
    Generate a WordCloud from the given text and save it as an image.

    Args:
        text (str): The input text for the WordCloud.
        output_filename (str): The filename to save the WordCloud image.
    """
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    wordcloud.to_file(output_filename)
    print(f"WordCloud saved as {output_filename}")
input_text = 'data science machine learning artificial intelligence'
output_file = 'wordcloud.png'
generate_wordcloud(input_text, output_file)
```
